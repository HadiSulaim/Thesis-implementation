```python
import cv2
import numpy as np
import os
from PIL import Image
import matplotlib.pyplot as plt

def load_image(path):
    image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print(f"Error: Could not load image at {path}")
    return image

def calculate_iou(mask1, mask2):
    intersection = np.logical_and(mask1, mask2)
    union = np.logical_or(mask1, mask2)
    iou = np.sum(intersection) / np.sum(union)
    return iou

def display_images(folder, title):
    # Get sorted list of image filenames
    filenames = sorted(os.listdir(folder))
    
    # Load the first and last images
    first_image_path = os.path.join(folder, filenames[0])
    last_image_path = os.path.join(folder, filenames[-1])
    
    first_image = Image.open(first_image_path)
    last_image = Image.open(last_image_path)
    
    # Display the images
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(first_image, cmap='gray')
    plt.title(f'{title} - First Image')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(last_image, cmap='gray')
    plt.title(f'{title} - Last Image')
    plt.axis('off')
    
    plt.show()

# Paths to the ground truth and segmented frames
gt_folder = r"C:\Users\user1\Desktop\Object_Tracking_Implementation\SAM2\segment-anything-2\notebooks\videos\str.bis58"
segmented_folder = r"C:\Users\user1\Desktop\Object_Tracking_Implementation\SAM2\segment-anything-2\notebooks\videos\All_Str.bis58\outputs_street"

# Display images from both folders
display_images(gt_folder, "Ground Truth")
display_images(segmented_folder, "Segmented")

# List of frame filenames
frame_filenames = sorted(os.listdir(gt_folder))

# Threshold for considering an object occluded
iou_threshold = 0.5

# Counter for occluded frames
occluded_frame_count = 0

for frame_filename in frame_filenames:
    gt_path = os.path.join(gt_folder, frame_filename)
    
    # Adjust the filename for the segmented frames
    frame_number = int(frame_filename.split('.')[0])  # Extract the frame number
    segmented_filename = f"frame_{frame_number:04d}.png"  # Match the segmented frame format
    segmented_path = os.path.join(segmented_folder, segmented_filename)

    # Load the ground truth and segmented masks
    gt_mask = load_image(gt_path)
    segmented_mask = load_image(segmented_path)

    # Ensure both masks are loaded
    if gt_mask is None or segmented_mask is None:
        continue

    # Resize the segmented mask to match the ground truth mask size
    if gt_mask.shape != segmented_mask.shape:
        segmented_mask = cv2.resize(segmented_mask, (gt_mask.shape[1], gt_mask.shape[0]))

    # Calculate IoU
    iou = calculate_iou(gt_mask > 0, segmented_mask > 0)
    print(f"Frame {frame_filename}: IoU = {iou:.2f}")  # Debugging output

    # Determine occlusion
    if iou < iou_threshold:
        occluded_frame_count += 1

print(f"Number of occluded frames: {occluded_frame_count}")
```
